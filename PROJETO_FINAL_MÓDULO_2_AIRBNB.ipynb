# **VEJA ABAIXO UM NOVO RELATRIO SEM OS OUTLIERS:**

```
df_clean = df.copy()
df_clean.drop(df_clean[df_clean.price > 200].index, axis=0, inplace=True)
df_clean.drop(df_clean[df_clean.minimum_nights > 4].index, axis=0, inplace=True)
df_clean.drop(df_clean[df_clean.reviews_per_month > 3].index, axis=0, inplace=True)

df_clean.drop('license', axis=1, inplace=True)
df_clean.drop('neighbourhood_group', axis=1, inplace=True)

df_clean.hist(bins=15, figsize=(15,10));
```

### **Q4. Qual a mdia dos preos de aluguel?**

```
# ver a mdia da coluna `price``
df_clean.price.mean()
```

### **Q4. Qual a correlao existente entre as variveis**

```
corr = df_clean[['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month',
    'calculated_host_listings_count', 'availability_365']].corr()

display(corr)
```
