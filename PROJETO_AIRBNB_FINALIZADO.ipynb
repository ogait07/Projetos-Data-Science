**OBSERVAÇÕES:**

* Os "pontinhos pretos" no gráfico acima, representam o momento em que o uso alcançou os valores demonstrados, quanto mais próximos eles estiverem uns dos outros, percebemos um padrão de uso social aceitável, porém, na medida que a bolinha vai se afastando das demais, percebemos um padrão **outlier**, que infelizmente, não representa um uso comum aceitável para ser usado na média de uso afim de gerarmos um relatório preciso.

* Na tabela **PRICE**, percebemos uma contagem geral (linha count) com o valor de 13296, porém, percebemos que 75% dos valores estão na casa do 115, o que demonstra uma discrepância nos números iniciais, e mais, **apenas 6.6% dos clientes gastaram valores acima de 200,00.**

* A contagem geral de **mínimos de noites (minimum_nights)** também se encontra na casa dos 13296 porém, 75% do uso foram para apenas 3, ou seja, passou a casa dos 365 dias por ano e 75% dos usuários usaram 3 noites. Observa também abaixo do gráfico acima que apensas 18.3% contrataram o serviço para além de 4 noites.

* na coluna **reviews_per_month** (avaliações por mês), notamos a classificação geral em 11046, (muita coisa né?), porém, quando acompanhamos os percentuais do mesmo no relatório, percebemos que a diferença real é muito grande em comparação com o valor principal, ou seja, **apenas 13.4% dos visitantes fazem mais de 4 avaliações por mês.**

* Essas médias são tiradas pela contagem de uso de imóveis nas regiões da cidade analisada, segue como algo muito comum a realidade em que pessoas prefiram uma região da cidade a outra, porém, conforme o uso do serviço, notamos uma diluição maior no acúmulo de informações sobre os imóveis nas regiões mais procuradas, demonstrando que o contrario (sobre as regiões com menor procura) também se torna proporcional sem a devida quantidade de amostras para comparar essas regiões de forma mais justa, estatísticamente falando. Em casos como esse citado acima, aproveitamos as informações que possuam uma maior segurança de amostragem para fazer uma análise mais justa e fiel sobre o objeto da pesquisa.

* São essas discrepância que chamamos de **outliers**, em casos como esses, seguindo o explicado acima, excluimos os mesmos e percebemos uma amostragem mais aceitável, que possui uma leitura mais segura e satisfatória.






# **VEJA ABAIXO UM NOVO RELATÓRIO SEM OS OUTLIERS:**

```
df_clean = df.copy()
df_clean.drop(df_clean[df_clean.price > 200].index, axis=0, inplace=True)
df_clean.drop(df_clean[df_clean.minimum_nights > 4].index, axis=0, inplace=True)
df_clean.drop(df_clean[df_clean.reviews_per_month > 3].index, axis=0, inplace=True)

df_clean.drop('license', axis=1, inplace=True)
df_clean.drop('neighbourhood_group', axis=1, inplace=True)

df_clean.hist(bins=15, figsize=(15,10));
```

### **Q4. Qual a média dos preços de aluguel?**

```
# ver a média da coluna `price``
df_clean.price.mean()
```

### **Q4. Qual a correlação existente entre as variáveis**

```
corr = df_clean[['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month',
    'calculated_host_listings_count', 'availability_365']].corr()

display(corr)
```
